```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../DataSets/Cat_human_New (1).csv")
data.head()
```

```python
for x in data.columns:
    if x == "label":
        break
    plt.scatter(data = data , x = x,y="label")
    plt.title(f" label vs {x}")
    plt.xlabel("Label")
    plt.ylabel(f"{x}")
    plt.show()
```
