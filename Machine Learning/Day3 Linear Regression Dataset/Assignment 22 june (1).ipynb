### Linear Regression
Task 1:
Use Linear_regresssion.csv file and apply linear regression to predict Y on basis of x
Evaluate through R2 method and check score of your predicted data with respect to Actual data during This process remember all the working cycle of ML (Data, preprocessing & cleaning, visualization, model, test, deployment )

```python
df = pd.read_csv("linear_reg.csv")
```

```python
df.dropna(axis=0)
df.head(13)
```

```python
x = np.array(df["X"]).reshape(-1, 1)
y=np.array(df["Y"]).reshape(-1, 1)
```

```python
#split test and train data
x_train,x_test ,y_train ,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
x_train
print("\nx_train",x_train)
print("\nx_train",x_test)
print("\nx_train",y_train)
print("\nx_train",y_test)
```

```python
#train model

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# R2 method and check score of your predicted data with respect to Actual data 

model.intercept_
```

```python
#model score
model.score(x_test,y_test)
```

```python
#model Coefcient
model.coef_
```

```python
#show us the predicted value

y_predicted = model.predict(x_test)
result  =pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_predicted.flatten()})
result
```

```python
#Error Calculation

mae = mean_absolute_error(y_test,y_predicted)
mse = mean_squared_error(y_test,y_predicted)
rmse = np.sqrt(mse)
print(f'Mean absolute error: {mae}')
print(f'Mean squared error: {mse}')
print(f'Root mean squared error: {rmse}')
```

```python
Act_minus_predict = sum((y_test - y_predicted)**2)
Act_minus_ActMean = sum((y_test - y_test.mean())**2)
r2 = 1 - Act_minus_predict/Act_minus_ActMean

print("r2 :" ,r2)
```

```python
data = pd.read_csv("insurance.csv")
#cleaning data
data = data.loc[:, ~data.columns.duplicated()]
data.dropna(axis =1 ,how='all')
data.dropna(axis=0, how='all')
data.drop_duplicates()

data
```

```python
df = data.drop(['charges'],axis = 1)
x = df
y =data['charges']
numeric = ['age','bmi','children']
categoric= ["region",'smoker','sex']

categoric
```

```python


numeric_transfomer = Pipeline(steps=[('scalar', StandardScaler())])
catego_transfomer = Pipeline(steps=[('encoder', OneHotEncoder())])


```

```python
preprocessor = ColumnTransformer(
transformers =[
    ('num',numeric_transfomer,numeric),
    ('cat',catego_transfomer,categoric)
])
```

```python
pipeline = Pipeline(steps=[
('preprocessor',preprocessor),
('classifier',LinearRegression())])
```

```python
pipeline.fit(x,y)
```

```python
joblib.dump(pipeline, 'linear.pkl')
```
